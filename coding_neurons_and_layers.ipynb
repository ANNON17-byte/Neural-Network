3 - Input Neuron

```python
inputs = [1, 2, 3]
weights = [0.2, 0.8, -0.5]
bias = 2  # bias is extra parameter added to weighed sum , which improve model learning and shift activation func

outputs = (inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias)

print(outputs)
```

Layer

```python
inputs = [1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],
 [0.5, -0.91, 0.26, -0.5],
 [-0.26, -0.27, 0.17, 0.87]]

weights1 = weights[0]
weights2 = weights[1]
weights3 = weights[2]
biases = [2, 3, 0.5]

bias1 = 2
bias2 = 3
bias3 = 0.5

output = [
    inputs[0]*weights1[0] + inputs[1]*weights1[1] + inputs[2]*weights1[2] + inputs[3]*weights1[3] + bias1,
    inputs[0]*weights2[0] + inputs[1]*weights2[1] + inputs[2]*weights2[2] + inputs[3]*weights2[3] + bias2,
    inputs[0]*weights3[0] + inputs[1]*weights3[1] + inputs[2]*weights3[2] + inputs[3]*weights3[3] + bias3,
]

print(output)
```

Layer Using Loops

```python
inputs = [1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],
 [0.5, -0.91, 0.26, -0.5],
 [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

output = []

#neuron_weight is single neuron weight
for neuron_weights, neuron_bias in zip(weights, biases):  #zip combines two list element by element
    neuron_output = 0
    for n_input, weight in zip(inputs, neuron_weights):
        neuron_output += n_input*weight
    neuron_output += neuron_bias # adding bias
    output.append(neuron_output)

print(output)

```

Single Neuron Using Numpy

```python

inputs = [1.0, 2.0, 3.0, 2.5]
weights = [0.2, 0.8, -0.5, 1.0]
bias = 2.0

import numpy as np

#converting to array for dot product
input_array = np.array(inputs)
weight_array = np.array(weights)

output = np.dot(input_array, weight_array) + bias

print(output)
```

Layers Of Neurons Using Numpy


```python
inputs = [1.0, 2.0, 3.0, 2.5]
weights = [[0.2, 0.8, -0.5, 1],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]
biases = [2.0, 3.0, 0.5]

import numpy as np

inpuut_array = np.array(inputs)
weight_array = np.array(weights)
bias_array = np.array(biases)

output = np.dot(weight_array, inpuut_array) + bias_array

print(output)
```

Layer Of Neurons Case 2 (whenn inp matrix is 3*4 and output matrix is 3*4)

```python
inputs = [[1.0, 2.0, 3.0, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, -0.8]]
weights = [[0.2, 0.8, -0.5, 1],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]
biases = [2.0, 3.0, 0.5]

import numpy as np

inpuut_array = np.array(inputs)
weight_array = np.array(weights)
bias_array = np.array(biases)

output = np.dot(inpuut_array, weight_array.T) + bias_array

print(output)
```

2 Layers

```python
inputs = [[1, 2, 3, 2.5],
          [2., 5., -1., 2],
          [-1.5, 2.7, 3.3, -0.8]]

weights = [[0.2, 0.8, -0.5, 1],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

weights2 = [[0.1, -0.14, 0.5],
            [-0.5, 0.12, -0.33],
            [-0.44, 0.73, -0.13]]

biases2 = [-1, 2, -0.5]

import numpy as np

inputs_array = np.array(inputs)
weights_array = np.array(weights)
biases_array = np.array(biases)
weights2_array = np.array(weights2)
biases2_array = np.array(biases2)

#first layer dot product

output1 = np.dot(inputs_array, weights_array.T) + biases_array

#second layer dot product

output2 = np.dot(output1, weights2_array.T) + biases2_array

print(output2)
```

Dense Layer Class


```python
import nnfs
from nnfs.datasets import spiral_data
import numpy as np

nnfs.init()

class dense_layer:

  def __init__(self, inputs, neurons):
    #randomly assigning weights
    self.weights = 0.01 * np.random.randn(inputs, neurons)
    #1 row and neurons columns
    self.biases = np.zeros((1, neurons))

  def forward(self, inputs):
    self.output = np.dot(inputs, self.weights) + self.biases

#create dataset
X,y = spiral_data(samples=100, classes =3)

#create dense layer with 2 inputs and 3 output values
dense = dense_layer(2,3)
dense.forward(X)

#output of first 5 data
print(dense.output[:5])
```

Activation Functions : Relu

ReLU introduces non-linear behavior, allowing networks to learn complex relationships

```python
import numpy as np
inputs = [0, 2, -1, 3.3, -2.7, 1.1, 2.2, -100]
output = np.maximum(0, inputs)
print(output)
```

```python
# ReLU activation class
class Activation_ReLU:
 # Forward pass
 def forward(self, inputs):
 # Calculate output values from input
  self.output = np.maximum(0, inputs)
```

Rectified Linear Unit

Explanation

Rectified → Negative values are removed (converted to 0)

Linear → Positive values remain linear (unchanged)

Unit → A single neuron operation

```python
import nnfs
from nnfs.datasets import spiral_data
import numpy as np

X,y = spiral_data(samples = 100, classes = 3)

#2 inputs and 3 outputs
dense = dense_layer(2,3)

#relu activation layer
activation1 = Activation_ReLU()

#forward pass
dense.forward(X)

activation1.forward(dense.output)

print(activation1.output[:5])
```

Softmax Activation Function : It converts a list of numbers into probabilities that sum to 1.

```python
inputs = [[1, 2, 3, 2.5],
 [2., 5., -1., 2],
 [-1.5, 2.7, 3.3, -0.8]]

# Get unnormalized probabilities
exp_values = np.exp(inputs - np.max(inputs, axis=1,keepdims=True))
 # Normalize them for each sample
probabilities = exp_values / np.sum(exp_values, axis=1,keepdims=True)
print(probabilities)
np.sum(probabilities, axis = 1)
```

```python
# axis = 0 for row and axis = 1 for columns
# keepdims = true for correct broadcasting, alligning matrix dimensions

class Activation_Softmax:

  def forward(self,input):

    exp_values = np.exp(inputs - np.max(inputs, axis = 1 , keepdims= True))

    probabilities = exp_values / np.sum(exp_values, axis = 1 , keepdims = True)

    self.output = probabilities
```
